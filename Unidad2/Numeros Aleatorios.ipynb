###Generaciòn de numeros aleatorios por el metodo de centros al cuadradado
####Alumno: Luis Tadeo Becerra Rocha


```python
# Parámetros del método de congruencia lineal
semilla = 2005  # Semilla inicial
a = 5          # Multiplicador
c = 3          # Incremento
m = 256        # Módulo

iteraciones = 100
resultados = []

for _ in range(iteraciones):
    # Aplicar la ecuación de congruencia lineal
    semilla = (a * semilla + c) % m
    resultados.append(semilla)

# Imprimir los resultados
for i, numero in enumerate(resultados, 1):
    print(f"Iteración {i}: {numero}")
```
