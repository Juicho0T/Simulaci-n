###Problema 1 Distribución Binomial

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución binomial
n_pedidos = 20  # Número total de pedidos diarios
p_exito = 0.7  # Probabilidad de éxito (satisfacer un pedido)
dias_semana = 7  # Simulación para una semana

# Generar la cantidad de pedidos satisfechos por día
pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, dias_semana)

# Mostrar los resultados
print("Pedidos satisfechos por día:", pedidos_satisfechos)

# Graficar los resultados
plt.bar(range(1, dias_semana + 1), pedidos_satisfechos, color='skyblue', edgecolor='black')
plt.xticks(range(1, dias_semana + 1))
plt.xlabel('Día')
plt.ylabel('Pedidos Satisfechos')
plt.title('Simulación de Pedidos Satisfechos por Día (Distribución Binomial)')
plt.show()
```
