###Problema 2 Distribución Exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial (tasa de llegada de vehículos)
lambda_llegada = 1 / 15  # En promedio, 1 llegada cada 15 minutos
horas_operacion = 12  # Duración de la simulación en horas
minutos_operacion = horas_operacion * 60  # Convertir a minutos

# Función para generar tiempos entre llegadas
def generar_tiempo_entre_llegadas(lambda_llegada):
    u = np.random.uniform(0, 1)  # Generar número aleatorio uniforme
    tiempo_entre_llegadas = -1 / lambda_llegada * np.log(1 - u)  # Aplicar transformada inversa
    return tiempo_entre_llegadas

# Generar tiempos entre llegadas de vehículos durante 12 horas
tiempos_entre_llegadas = []
tiempo_actual = 0
while tiempo_actual < minutos_operacion:
    tiempo_nuevo = generar_tiempo_entre_llegadas(lambda_llegada)
    tiempo_actual += tiempo_nuevo
    if tiempo_actual < minutos_operacion:
        tiempos_entre_llegadas.append(tiempo_nuevo)

# Mostrar los tiempos entre llegadas generados
print("Tiempos entre llegadas generados:", tiempos_entre_llegadas)

# Graficar la distribución de los tiempos entre llegadas
plt.hist(tiempos_entre_llegadas, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos entre Llegadas (Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
