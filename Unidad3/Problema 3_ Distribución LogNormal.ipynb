###Problema 3 Distribución LogNormal

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución lognormal
mu = np.log(10)  # Media logarítmica (aproximada)
sigma = 2  # Desviación estándar en la escala logarítmica
num_clientes = 50  # Número de clientes en la simulación

# Generar tiempos de espera para los clientes
tiempos_espera = np.random.lognormal(mean=mu, sigma=sigma, size=num_clientes)

# Mostrar los tiempos de espera generados
print("Tiempos de espera generados:", tiempos_espera)

# Graficar la distribución de los tiempos de espera
plt.hist(tiempos_espera, bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de Tiempos de Espera en el Mostrador (LogNormal)')
plt.xlabel('Tiempo de espera (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
